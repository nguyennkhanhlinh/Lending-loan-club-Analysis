- Does not meet the credit policy. Status:Charged Off: Không đáp ứng chính sách tín dụng và bị xóa sổ.

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
X = data_final.drop('loan_status', axis=1)  
y = data_final['loan_status']  
```

```python
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)
```

```python
print(y_resampled.value_counts())
```

# Mô hình Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Tách dữ liệu
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=102)

# Huấn luyện mô hình
model = LogisticRegression()
model.fit(X_train, y_train)
model.coef_
```

```python
# Đánh giá kết quả mô hình
predictions = model.predict(X_test)
print('Accuracy score:', accuracy_score(y_test, predictions))
print('Confusion matrix:')
print(confusion_matrix(y_test, predictions))
```

Confusion matrix

```python
class_names=[0,1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

# vẽ heatmap
cnf_matrix=confusion_matrix(y_test, predictions)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

- True Positive (TP): 12 018: số khách hàng thực tế là trả được nợ và được mô hình dự đoán đúng là trả được nợ
- False Positive (FP): 696: số khách hàng thực tế là trả được nợ nhưng mô hình lại dự đoán là không trả được nợ

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

- Precision : Precision cho lớp 0 là 0.85, và cho lớp 1 là 0.94. Điều này có nghĩa là khi mô hình dự đoán một mẫu thuộc lớp 1, có 94% khả năng dự đoán đó là chính xác, trong khi khi dự đoán lớp 0, tỷ lệ chính xác thấp hơn (85%).
- Recall (Độ nhạy): Recall cho lớp 0 là 0.95, và cho lớp 1 là 0.80. Điều này có nghĩa là mô hình có khả năng nhận diện lớp 0 tốt hơn lớp 1. Cụ thể, 95% các mẫu thuộc lớp 0 đã được mô hình nhận diện đúng, nhưng chỉ có 84% các mẫu lớp 1 được nhận diện chính xác.
- F1-Score:F1-Score cho cả hai lớp đều khá cao (0.9 cho lớp 0 và 0.89 cho lớp 1), cho thấy mô hình đạt được sự cân bằng tốt giữa precision và recall.
- Accuracy (Độ chính xác chung):Accuracy đạt 0.89, tức là 89% tổng số mẫu được phân loại đúng

```python
from sklearn import metrics

y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.metrics import recall_score, precision_score, accuracy_score, f1_score
```

```python
# Chọn ngưỡng cắt
thresholds = np.linspace(0.1, 0.9, 9)

# List để lưu trữ kết quả đánh giá
evaluation_scores = []

# Duyệt qua từng ngưỡng
for threshold in thresholds:
    # Dự đoán nhãn với ngưỡng hiện tại
    predicted_labels = ((model.predict_proba(X_test)[:, 1] >= threshold)).astype(int)

    # Tính các metric: recall, precision, accuracy
    recall = recall_score(y_test, predicted_labels)
    precision = precision_score(y_test, predicted_labels)
    accuracy = accuracy_score(y_test, predicted_labels)
    f1 = f1_score(y_test, predicted_labels)
    # Thêm kết quả vào list
    evaluation_scores.append([threshold, recall, precision, accuracy, f1])

# Tạo DataFrame từ list kết quả
df_ = pd.DataFrame(evaluation_scores, columns=['Threshold', 'Recall', 'Precision', 'Accuracy', 'F1 Score'])

# In ra bảng kết quả
print(df_)
```

Với mục tiêu là cân bằng tốt giữa phát hiện đủ các khoản vay “rủi ro” và không quá tốn kém cho những cảnh báo nhầm (Precision), đồng thời tối ưu chung F1-Score và Accuracy, thì ngưỡng 0.4 là lựa chọn hợp lý:

+ F1-Score cao nhất (0.89) tại ngưỡng 0.4, nghĩa là mô hình có sự cân bằng tốt nhất giữa Precision và Recall ở điểm này.
+ Accuracy lớn nhất (0.89) cũng đạt tại ngưỡng 0.4, tức tỷ lệ phân loại đúng tổng thể tối ưu.
- Tại ngưỡng này:
+ Recall = 90.85% → Phát hiện 90.85% khoản vay rủi ro (giảm bỏ sót).
+ Precision = 87.28% → Trong số những khoản vay đánh dấu rủi ro, có 87.28% thực sự rủi ro (hạn chế cảnh báo nhầm).
- Nếu hạ threshold thấp hơn (ví dụ 0.3), tăng recall (0.9) nhưng giảm precision (0.86) –> nhiều cảnh báo nhầm hơn.
- Nếu nâng threshold cao hơn (ví dụ 0.5), tăng precision (0.93) nhưng giảm recall (0.83) –> bỏ sót nhiều khoản rủi ro hơn.
=> Ngưỡng 0.4 vừa giúp phát hiện hầu hết khoản vay xấu, đồng thời không lãng phí quá nhiều nguồn lực điều tra các khoản vay tốt.

```python
import numpy as np
import joblib
input_data = (
    50000.0,     # loan_amnt
    6.5,         # int_rate
    70000.0,     # annual_inc
    15.0,        # dti
    40.0,        # revol_util
    1.0,         # chargeoff_within_12_mths
    2.0,         # collections_12_mths_ex_med
    5.0,         # emp_length
    1.0,         # collection_recovery_fee
    5.0,         # home_ownership
    1.0,         # sub_grade
    0.0,         # verification_status
    0.0,         # acc_now_delinq
    10.0,        # delinq_2yrs
    2.0,         # acc_open_past_24mths
    1.0,         # mort_acc
    0.0,         # recoveries
    650.0        # last_fico_range_avg
)

input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Load mô hình đã huấn luyện
loaded_model = joblib.load('model_1.sav')

# Dự đoán
prediction = loaded_model.predict(input_data_reshaped)

# In kết quả
if prediction[0] == 0:
    print('The person is eligible to pay for the loan')
else:
    print('The person is not eligible to pay for the loan')
```
